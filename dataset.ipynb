 Using R or Python, write a program to read any dataset, show the number of
samples and number of features in dataset. Select any one feature from the
same dataset and calculate mean and median values for it. Plot Histogram for
the same.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read dataset
df = pd.read_csv(r"c:\Users\HP\Downloads\archive\Titanic-Dataset.csv")

# 2. Show number of samples (rows) and features (columns)
print("Number of samples (rows):", df.shape[0])
print("Number of features (columns):", df.shape[1])

# 3. Select any one feature (example: first column)
feature = df.columns[5]
print("Selected feature:", feature)

mean_value = df[feature].mean()
median_value = df[feature].median()

print("Mean of feature:", mean_value)
print("Median of feature:", median_value)

# 4. Plot histogram
plt.hist(df[feature], bins=20)
plt.title(f"Histogram of {feature}")
plt.xlabel("Age")
plt.ylabel("No. of Persons")
plt.show()
```
