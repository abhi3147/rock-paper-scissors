Using R or Python, write a function to calculate Euclidean distance between
any two vectors. Show the results of if vect1 is (2, 4, 4, 7) and vect2 is (1, 2,
2, 10).

```python
import math

def euclidean_distance(v1, v2):
    return math.sqrt(sum((a - b)**2 for a, b in zip(v1, v2)))

vect1 = [2, 4, 4, 7]
vect2 = [1, 2, 2, 10]

distance = euclidean_distance(vect1, vect2)
print("Euclidean Distance:", distance)
```
