Dynamic Programming: Implement Python function for computing Fibonacci
number and factorial of a number using Dynamic Programming.

```python
# Fibonacci using memoization (top-down DP)
fib_memo = {}

def fibonacci_memo(n):
    if n <= 1:
        return n
    if n not in fib_memo:
        fib_memo[n] = fibonacci_memo(n-1) + fibonacci_memo(n-2)
    return fib_memo[n]

# Example
print("Fibonacci(10):", fibonacci_memo(10))

fib = {}

def fibonacci(n):
    if n <=1:
        return n
    if n not in fib:
        fib[n] = fibonacci(n-1) + fibonacci(n-2)
    return fib[n]
print(fibonacci(10))
```

```python
# Factorial using bottom-up DP
def factorial_dp(n):
    dp = [1] * (n + 1)   # dp[0] = 1

    for i in range(1, n + 1):
        dp[i] = dp[i - 1] * i

    return dp[n]

# Example
print("Factorial(6):", factorial_dp(6))

def factorial(n):
    fac = [1] * (n + 1)

    for i in range(1, n+1):
        fac[i] = fac[i - 1] * i
    
    return fac[n]
```
