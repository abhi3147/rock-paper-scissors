 Using R or Python, write a function to calculate Manhattan distance between
any two vectors. Show the results of if vect1 is (3, 6, 8, 9) and vect2 is (1, 7,
8, 10).

```python
def manhattan_distance(v1, v2):
    return sum(abs(a - b) for a, b in zip(v1, v2))

vect1 = [3, 6, 8, 9]
vect2 = [1, 7, 8, 10]

distance = manhattan_distance(vect1, vect2)
print("Manhattan Distance:", distance)
```
