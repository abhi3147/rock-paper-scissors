```python
# def greedy_knapsack_01(weights, values, capacity):
#     n = len(values)

#     # Calculate value-to-weight ratios
#     ratio = [(values[i] / weights[i], weights[i], values[i]) for i in range(n)]

#     # Sort by ratio in descending order
#     ratio.sort(reverse=True)

#     total_value = 0
#     total_weight = 0
#     selected_items = []

#     for r, wt, val in ratio:
#         if total_weight + wt <= capacity:
#             selected_items.append((wt, val, r))
#             total_weight += wt
#             total_value += val

#     return total_value, total_weight, selected_items


# # Example
# weights = [2, 3, 4, 5]
# values  = [3, 4, 5, 8]
# capacity = 5


def greedy(weights,values,capacity):
    n = len(values)
    ratio = [(values[i]/weights[i],weights[i],values[i]) for i in range(n)]
    ratio.sort(reverse=True)

    total_wt = 0
    total_val = 0
    selected = []
    for r,wt,val in ratio:
        if total_wt + wt <= capacity:
            total_wt += wt
            total_val += val
            selected.append((val,wt))

    return total_val,total_wt,selected
weights = [2, 3, 4, 3]
values  = [3, 4, 5, 8]
capacity = 5

max_value, used_weight, items = greedy(weights, values, capacity)

print("Total Value (Greedy):", max_value)
print("Total Weight used:", used_weight)
print("Items selected (value, weight):",items)

```
